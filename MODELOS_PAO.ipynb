fancy way to fill Nulls: K-nearest Neihbour imputation

You can select a sample with missing values and find the nearest neighbours utilising some kind of distance metric, for example, Euclidean distance. Then you can take the mean of all nearest neighbours and fill up the missing value. You can use the KNN imputer implementation for filling missing values like this.

How it works? 

1. Select the K number of the neighbours
2. Calculate the euclidean distance of K number of neighbors
3. Among these k neighbors, count the number of data points in each category
4. Assign the new data points to that category for which the number of the neighbours is maximum
