# II. Limpieza de data

### Elegir datos

```python
# COMBINACIÓN
kaggle_score = 0
combinacion_general = ''
```

```python
# SIN predictions Kaggle
# Agregar -1 al target del test
data_test_target = data_test.copy()
data_test_target['target'] = -1

# Dataset combinado que será limpiado
data_clean = pd.concat([data, data_test_target], axis = 0).reset_index(drop = True)
print(data_clean.shape)
```

```python
data_clean.head()
```

```python
# CON predictions Kaggle
test_targets = pd.read_csv('preds_Stacking_XGB_SVC_city_TDummies_MinMax_SMOTE6.csv')
test_targets = pd.concat([data_test, test_targets['target']], axis=1)

data_test_target = data_test.copy()
data_test_target['target'] = -1

# Dataset combinado que será limpiado
data_clean = pd.concat([data, test_targets], axis = 0).reset_index(drop = True)
data_clean = pd.concat([data_clean, data_test_target]).reset_index(drop = True)
print(data_clean.shape)

combinacion_general = combinacion_general + '_conpred'
```

```python
# Dropear columnas irrelevantes
data_clean = data_clean.drop(['gender'], axis=1)

if ('city' in list(data_clean.columns)) == False:
    combinacion_general = combinacion_general + '_' + 'NOcity'
if ('gender' in list(data_clean.columns)) == True:
    combinacion_general = combinacion_general + '_' + 'gender'  

# Cambiar id's a index
data_clean = data_clean.set_index("enrollee_id")

# Cambiar caracteres extraños (para XGB)
data_clean['experience'] = data_clean['experience'].str.replace('<1', '0')
data_clean['experience'] = data_clean['experience'].str.replace('>20', '24')
data_clean['company_size'] = data_clean['company_size'].str.replace('<10', '0')
data_clean['last_new_job'] = data_clean['last_new_job'].str.replace('>4', '5')

# Redondear valores largos en city development index
data_clean['city_development_index'] = data_clean['city_development_index'].round(3)
combinacion_general
```

#### ¿Incluiré `city`?

```python
# Dropear filas con ciudades que no están en el data_test
ciudades_test = data_test['city'].unique().tolist()
filtro = data_clean['city'].isin(ciudades_test)
data_clean = data_clean[filtro]

if ('city' in list(data_clean.columns)) == True:
    combinacion_general = combinacion_general + '_' + 'city'
```

#### ¿Transformaré nulos?

```python
# Exp y last_job a numéricas
data_clean['experience'] = data_clean['experience'].astype('float')
data_clean['last_new_job'] = data_clean['last_new_job'].str.replace('never', '0')
data_clean['last_new_job'] = data_clean['last_new_job'].astype('float')

if data_clean['experience'].dtype == 'float64':
    combinacion_general = combinacion_general + '_' + 'NumExp'
    
if data_clean['last_new_job'].dtype == 'float64':
    combinacion_general = combinacion_general + '_' + 'NumLastJob'
```

```python
# Llenar NaN's en variables numéricas | Mean

data_clean['experience'] = data_clean['experience'].fillna(round(data_clean['experience'].mean()))
data_clean['last_new_job'] = data_clean['last_new_job'].fillna(round(data_clean['last_new_job'].mean()))
combinacion_general = combinacion_general + 'Media'

#Llenar NaN's en variables categóricas | Moda
for column in data_clean.columns:
    if data_clean[column].dtype == 'object':
        mode = str(data_clean[column].mode()[0])
    
    else:
        mode = float(data_clean[column].mean()[0])
    
    data_clean[column] = data_clean[column].fillna(mode)
    
combinacion_general = combinacion_general + 'Moda'
```

### Dummies

```python
# Dummies de TODO (ya no es necesario hacer mapping)
data_clean = pd.get_dummies(data=data_clean, drop_first=True)
combinacion_general = combinacion_general + '_' + 'TDummies'

data_clean.head(4)
```

#### ¿Scaling?

```python
# Scaler | MinMax
scaler = MinMaxScaler(feature_range=(0,1))
X = scaler.fit_transform(data_clean.drop("target", axis=1))

dtf_scaled= pd.DataFrame(X, columns=data_clean.drop("target", axis=1).columns, index=data_clean.index)
dtf_scaled["target"] = data_clean["target"]
data_clean = dtf_scaled.copy()

combinacion_general = combinacion_general + '_' + 'MinMax'

# KNEIGHBORS - USARSE CON MINMAX y COLUMNAS NUMÉRICAS (EXP Y LAST_NEW_JOB)
#imputer = KNNImputer(n_neighbors=5)

#data_clean = pd.DataFrame(imputer.fit_transform(data_clean),columns = data_clean.columns)

#combinacion_general = combinacion_general + '_' + 'KNeigh'
```

### Separar datasets

```python
# Separar datasets
data_clean_test = data_clean[data_clean['target'] == -1]
data_clean_test = data_clean_test.drop('target', axis=1)
data_clean = data_clean[data_clean['target'] != -1]
```

```python
# !! EJECUTAR !!
X = data_clean.drop(['target'], axis=1)
y = data_clean.target
```

```python
X.shape
```

### Balanceo

```python
# SMOTE
k_number = 6
smote = SMOTE(random_state=1, k_neighbors=k_number) #Ajustar k_neighbors=5 (default)
X, y = smote.fit_resample(X, y)

combinacion_general = combinacion_general + '_' + f'SMOTE{k_number}'
```

```python
# BSMOTE
k_number = 5
smote = BorderlineSMOTE(random_state=1, k_neighbors=k_number) #Ajustar k_neighbors=5 (default)
X, y = smote.fit_resample(X, y)

combinacion_general = combinacion_general + '_' + f'BSMOTE{k_number}'
```

```python
# SVMSMOTE
k_number = 5
smote = SVMSMOTE(random_state=1, k_neighbors=k_number) #Ajustar k_neighbors=5 (default)
X, y = smote.fit_resample(X, y)

combinacion_general = combinacion_general + '_' + f'SVSMOTE{k_number}'
```

```python
# ADASYN
smote = ADASYN(random_state=1)
X, y = smote.fit_resample(X, y)

combinacion_general = combinacion_general + '_' + f'ADASYN{k_number}'
```

### Dividir datasets

```python
# DIVIDIR DATASETS
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =1)
print(f'Combinacion: {combinacion_general}')
```

```python
param_grid = [{'n_estimators': [100], 'max_features': [6] , 'max_depth': [40, 60, 80, 100], 'min_samples_leaf': [5]}]
model = RandomForestClassifier(random_state=1)

grid_search = GridSearchCV(model, param_grid, cv=5, scoring='roc_auc', return_train_score=True)
grid_search.fit(X, y)

grid_search.best_params_
```

# VI. AUC Analysis

```python
# !!!EJECUTAR SOLO UNA VEZ POR TRIAL!!! DataFrame con scores de AUC
modelos = ['Logistic', 'XGB', 'SVC', 'Linear SVC', 'Decision Tree', 'Random Forest',\
           'Linear Discriminant', 'NN', 'Stacking']
scores = [0 for i in range(0, len(modelos))]
aucs = {'Modelo':modelos}
aucs = pd.DataFrame(aucs)
```

```python
balancers = ['SMOTE', 'BSMOTE', 'SVMSMOTE', 'ADASYN']

for balancer in balancers:
    combinacion = f'{combinacion_general}_{balancer}'
    aucs[combinacion] = [0 for i in range(0, len(modelos))]
    
    ### Definir variables
    X = data_clean.drop(['target'], axis=1)
    y = data_clean.target
    
    ### BALANCEO
    sampling_size = 0.9
    
    if balancer == 'SMOTE':
        smote = SMOTE(random_state=1) #Ajustar k_neighbors=5 (default)
        X, y = smote.fit_resample(X, y)
    
    elif balancer == 'BSMOTE':
        smote = BorderlineSMOTE(random_state=1)
        X, y = smote.fit_resample(X, y)

    elif balancer == 'SVMSMOTE':
        smote = SVMSMOTE(random_state=1)
        X, y = smote.fit_resample(X, y)
    
    elif balancer == 'ADASYN':
        smote = ADASYN(random_state=1)
        X, y = smote.fit_resample(X, y)
    
    
    ### DIVIDIR DATASETS
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state =1)
    print(f'\n----{balancer}----')
    
    #### LOGISTIC REGRESSION
    # Entrenar modelo
    model = LogisticRegression(max_iter=10_000).fit(X_train, y_train) # solver='lbfgs' 'liblinear'

    # Hacer predicciones
    y_pred_test = model.predict(X_test)

    # Metrics 
    print(f'Logistic | AUC: {metrics.roc_auc_score(y_test, y_pred_test)}')

    # DataFrame AUC's
    score = metrics.roc_auc_score(y_test, y_pred_test)
    aucs.loc[aucs['Modelo'] == 'Logistic', combinacion] = score


    #### XGBCLASSIFIER
    model = XGBClassifier()
    model.fit(X_train, y_train)

    y_pred_test = model.predict(X_test)

    # Metrics 
    print(f'XGB | AUC: {metrics.roc_auc_score(y_test, y_pred_test)}')

    # DataFrame AUC's
    score = metrics.roc_auc_score(y_test, y_pred_test)
    aucs.loc[aucs['Modelo'] == 'XGB', combinacion] = score


    #### SVC
    model = SVC()
    model.fit(X_train, y_train)

    y_pred_test = model.predict(X_test)

    # Metrics 
    print(f'SVC | AUC: {metrics.roc_auc_score(y_test, y_pred_test)}')

    # DataFrame AUC's
    score = metrics.roc_auc_score(y_test, y_pred_test)
    aucs.loc[aucs['Modelo'] == 'SVC', combinacion] = score


    #### LINEAR SVC
    model = LinearSVC(max_iter=5_000)
    model.fit(X_train, y_train)

    y_pred_test = model.predict(X_test)

    # Metrics 
    print(f'LSVC | AUC: {metrics.roc_auc_score(y_test, y_pred_test)}')

    # DataFrame AUC's
    score = metrics.roc_auc_score(y_test, y_pred_test)
    aucs.loc[aucs['Modelo'] == 'Linear SVC', combinacion] = score

    
    ### DECISION TREE
    model = DecisionTreeClassifier()
    cv = RepeatedStratifiedKFold(random_state=1)

    # Metrics 
    scores = cross_val_score(model, X, y, scoring='roc_auc', cv=cv, n_jobs=-1)
    print('D Tree | AUC: %.3f' % mean(scores))

    # DataFrame AUC's
    score = mean(scores)
    aucs.loc[aucs['Modelo'] == 'Decision Tree', combinacion] = score

    
    ### RANDOM FOREST
    model = RandomForestClassifier(random_state=1)
    model.fit(X_train, y_train)

    y_pred_test = model.predict(X_test)

    # Metrics 
    print(f'R Forest | AUC: {metrics.roc_auc_score(y_test, y_pred_test)}')

    # DataFrame AUC's
    score = metrics.roc_auc_score(y_test, y_pred_test)
    aucs.loc[aucs['Modelo'] == 'Random Forest', combinacion] = score

    
    ### LINEAR DISCRIMINATION
    model = LinearDiscriminantAnalysis()
    model.fit(X_train, y_train)

    y_pred_test = model.predict(X_test)

    # Metrics 
    print(f'L Discrimination | AUC: {metrics.roc_auc_score(y_test, y_pred_test)}')

    # DataFrame AUC's
    score = metrics.roc_auc_score(y_test, y_pred_test)
    aucs.loc[aucs['Modelo'] == 'Linear Discriminant', combinacion] = score

    
    ### NN
    model = MLPClassifier(random_state=1)
    model.fit(X_train, y_train)

    y_pred_test = model.predict(X_test)

    # Metrics 
    print(f'NN | AUC: {metrics.roc_auc_score(y_test, y_pred_test)}')

    # DataFrame AUC's
    score = metrics.roc_auc_score(y_test, y_pred_test)
    aucs.loc[aucs['Modelo'] == 'NN', combinacion] = score  
    
    
    ###STACKING CLASSIFIER
    estimators_m = [('rf',RandomForestClassifier(random_state=1)),('log',LogisticRegression(max_iter=10_000))]
    model = StackingClassifier(estimators=estimators_m, final_estimator=LogisticRegression())
    model.fit(X_train, y_train)

    y_pred_test = model.predict(X_test)

    # Metrics 
    print(f'Stacking | AUC: {metrics.roc_auc_score(y_test, y_pred_test)}')
    
    # DataFrame AUC's
    score = metrics.roc_auc_score(y_test, y_pred_test)
    aucs.loc[aucs['Modelo'] == 'Stacking', combinacion] = score
    
sns.heatmap(aucs.iloc[:, 1:])
```
